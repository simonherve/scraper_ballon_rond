```python
import pandas as pd
pd.set_option('display.max_rows', 800)
from bs4 import BeautifulSoup
import json
import warnings
warnings.filterwarnings("ignore")
```

```python
def get_team(data):
    
    tab = []
    for elem in ["home", "away"]:
        tab.append(data["matchCentreData"][elem]["name"])
    
    return tab

def parse_file(x):
    x = json.loads(x)
    return get_team(x)
```

```python
df = pd.read_csv("test.csv", index_col=0)
```

```python
#df = pd.read_csv("./raw_data/data_L1_2018.csv", index_col=0)
df["return"] = df["data"].apply(parse_file)
df[["HomeTeam", "AwayTeam"]] = pd.DataFrame(df["return"].tolist(), index=df.index)
df.drop("return", axis=1, inplace=True)
df.shape
```
